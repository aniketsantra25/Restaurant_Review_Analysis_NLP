```python
import pandas as pd
import re
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
nltk.download('wordnet')
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier  
```

```python
stop_words = stopwords.words("english")
```

```python
ds = pd.read_csv("Restaurant_Reviews.tsv", delimiter="\t")
```

```python
ds
```

```python
reviews = ds['Review']
```

```python
reviews
```

```python
ps = PorterStemmer()
lm = WordNetLemmatizer()
```

```python
corpus = []
```

```python
for i in reviews:
    review = re.sub('[^a-zA-Z]', ' ', i)
    review = review.lower()
    words = review.split()
    stem_word = []
    for word in words:
        if word not in stop_words:
            x = ps.stem(word)
            stem_word.append(x)
    review = ' '.join(stem_word)
    corpus.append(review)
```

```python
corpus
```

```python
bag = CountVectorizer(max_features = 1500)
```

```python
X = bag.fit_transform(corpus)
```

```python
X = X.toarray()
```

```python
X.shape
```

```python
X
```

```python
y = ds['Liked'].values
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 49)
```

```python
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
model = Sequential()
```

```python
model.add(Dense(input_shape=(1500,), units = 64, activation = 'relu'))
```

```python
model.add(Dense(units = 32, activation = 'relu'))
```

```python
model.add(Dense(units = 8, activation = 'relu'))
```

```python
model.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = Adam(learning_rate = 0.000001), metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs = 1000)
```

```python
y_pred_nn = model.predict(X_test)
```
